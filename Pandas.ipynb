```python
df2=pd.DataFrame({'Address':['A0','A1','A2','A3','A4','A5'],
                 'Status':['B0','B1','B2','B3','B4','B5'],
                'Role':['C0','C1','C2','C3','C4','C5'],
                'Salary':['D0','D1','D2','D3','D4','D5']})
```

```python
df2
```
